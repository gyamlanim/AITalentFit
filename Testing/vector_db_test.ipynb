# Vector Database: PDF Resume Add Test

Test for adding a resume to the database using a PDF file and `enter_resume_to_db`.

```python
import os
import sys
from custom_rag import CustomRAG
from PyPDF2 import PdfReader

os.chdir(r"c:\Users\egyamah\ResumeParsing")
if os.getcwd() not in sys.path:
    sys.path.insert(0, os.getcwd())

test_db_name = "persistent_vector_test.db"
rag = CustomRAG(test_db_name)
```

```python
try:
    with open("data/sample_resumes/ArjunResume-AIGEN.pdf", 'rb') as f:
        reader = PdfReader(f)
        resume_text = "".join(page.extract_text() for page in reader.pages)
    result = rag.enter_resume_to_db(resume_text, {"filename": "ArjunResume-AIGEN.pdf"})
    if isinstance(result, str) and result.startswith("Error"):
        print(result)
except Exception as e:
    import traceback
    print(f"Error type: {type(e).__name__}")
    print(f"Error storing resume: {e}")
    traceback.print_exc()
```

```python
try:
    job_description = "Python developer web development"
    results = rag.retrieve_top_resumes(job_description, top_n=2)
    for i, result in enumerate(results, 1):
        name = result['metadata'].get('name', 'Unknown')
        similarity = result['similarity']
        print(f"{i}. {name} (Similarity: {similarity:.3f})")
except Exception as e:
    import traceback
    print(f"Error type: {type(e).__name__}")
    print(f"Error in RAG search: {e}")
    traceback.print_exc()
```
